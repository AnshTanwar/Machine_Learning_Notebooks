# <div style="background-color:#fff1cc; padding: 20px;">Let us use ydata-profiling</div>

```python
import pandas as pd
from ydata_profiling import ProfileReport

df = pd.read_csv("/kaggle/input/list-of-startups/Listofstartups.csv", encoding='ISO-8859-1')
df.head()
```

```python
profile = ProfileReport(df,title="list-of-startups: Profiling Report", explorative=True,)
profile.to_file("output1.html")
```
